# Neural Network on Iris dataset

Building the model

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
X_train, X_test, Y_train, Y_test = train_test_split(iris.data, iris.target, test_size=0.2)
```

```python
import tensorflow as tf

model = tf.keras.Sequential([tf.keras.layers.Dense(10, activation='relu', input_shape=(4,)),
         tf.keras.layers.Dense(3, activation='softmax')])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(
    X_train,Y_train,
    epochs=50,
    batch_size=8,
    validation_data=(X_test,Y_test)
)
```

```python
import numpy as np

model.evaluate(X_test,Y_test)
predictions = model.predict(X_test)
print(predictions[0])
print("model prediction :", np.argmax(predictions[0]))
print("expected result :", Y_test[0])
```

```python
def predict(SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm):
  X_ = np.array([[SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm]])
  predictions = model.predict(X_)
  return np.argmax(predictions[0])

test = predict(6.2, 2.6, 4., 1.5)
print(test)
```

Automatisation

```python
import pandas as pd

y_pred = np.argmax(model.predict(X_test), axis=1)

data = {
    'y_true' : Y_test,
    'y_pred' : y_pred
}

df = pd.DataFrame(data)
df.to_csv('resultats_iris.csv', index=False)
df.head()
```
